# Content Based Recommendation system

recommendations based on the movie’s plot summaries given in the overview column. 
So if user gives a movie title, goal is to recommend movies that share similar plot summaries.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
movies_clean.head()['overview']
```

For recommendation engine we need to create a vector of matrix because recommendation system is based on pairwise similarity so we represent every movie in terms of vectors

"Overview" is a string column and model cannot understand a sentence
For this we use NLP concept "tfidf"

Tfidf technique helps us to covert sentences into document matrix

min_df=3: ignore the terms that have frequency less than threshold value=3
max_features=none:

strip_accent='unicode': Remove accents and perform other character normalization;‘ascii’ is a fast method that only works on characters that have an direct ASCII mapping. ‘unicode’ is a slightly slower method that works on any characters.

analyzer='word': Whether the feature should be made of word or character n-grams

token_pattern=r'\w{1,}': remove unnecessary punctuations

ngram_range=(1, 3): lower and upper boundary of the range of n-values- take combination of 1 to 3 different kind of words

stop_words = 'english': remove all unnecessary characters like "the,at,of,is,etc." since these chars exists multiple number of times

```python
tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
```

Filling NaNs with empty string

```python
movies_clean['overview'] = movies_clean['overview'].fillna('')
```

Fitting the TF-IDF on the 'overview' text

Fit_transform converts it into a sparse matrix. Sparse matrix has many 0 and less non-zero values

```python
tfv_matrix = tfv.fit_transform(movies_clean['overview'])
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

So we have 4803 records and 10417 different features. Number of features depend upon ngram. Combination of different words become our columns

```python
from sklearn.metrics.pairwise import sigmoid_kernel
```

we use sigmoid to transform our input between 0 & 1. This Sigmoid is also used in logistic regression. Purpose is to check how overview1 is related to overview2

To Compute the sigmoid kernel use tfv_matrix twice cz only then the combination of matrix can be done. Everything will be based on the vectors

```python
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
indices = pd.Series(movies_clean.index, index=movies_clean['original_title']).drop_duplicates()
```

```python
def give_rec(title, sig=sig):
    idx = indices[title] # Get the index corresponding to original_title
    sig_scores = list(enumerate(sig[idx])) # Get the pairwsie similarity scores 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True) # Sort the movies 
    sig_scores = sig_scores[1:11] # Scores of the 10 most similar movies
    movie_indices = [i[0] for i in sig_scores]     # Movie indices
    return movies_clean['original_title'].iloc[movie_indices]  # Top 10 most similar movies
```
